```python
import pandas as pd

# Charger les données
data = pd.read_csv('Marathon du web ODDO 2023.csv', sep=';')

# Vérifier les premières lignes pour comprendre la structure des données
print(data.head())

# Remplacer les virgules par des points pour la conversion des valeurs
data['Nombre de voyages'] = data['Nombre de voyages'].str.replace(',', '.').astype(float)
data['Poids de voyages'] = data['Poids de voyages'].str.replace(',', '.').astype(float)


# Vérifier les types de données
print(data.info())

# Vérifier les valeurs manquantes
print(data.isnull().sum())

# Calculer les sommes des "Nombre de voyages" et "Poids de voyages"
total_voyages = data['Nombre de voyages'].sum()
total_poids = data['Poids de voyages'].sum()

# Calcul des ratios
data['Ratio voyages'] = data['Nombre de voyages'] / total_voyages * 100
data['Ratio poids'] = data['Poids de voyages'] / total_poids * 100

# Vérifier si les ratios totalisent 100 % ou 1
print(f"Somme des ratios des voyages : {data['Ratio voyages'].sum()}")
print(f"Somme des ratios du poids des voyages : {data['Ratio poids'].sum()}")
```

```python
# Calculer les statistiques descriptives sur "Nombre de voyages" et "Poids de voyages"
print(data[['Nombre de voyages', 'Poids de voyages']].describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Exemple de structure de DataFrame
data = {
    'Categorie': ['TOTAL', 'Jeunes', 'Fréquents', 'Occasionnels', 'WE 1€'],
    '2023_Nombre de voyages': [24891573, 8241046, 9830515, 13508564, 707381],
    '2024_Nombre de voyages': [26992719, 9097637, 9155030, 14587668, 847604],
}

df = pd.DataFrame(data)

# Tracer le nombre de voyages en 2023 et 2024 pour chaque catégorie
df.set_index('Categorie')[['2023_Nombre de voyages', '2024_Nombre de voyages']].plot(kind='bar', figsize=(10,6))
plt.title("Nombre de voyages en 2023 et 2024 par catégorie")
plt.ylabel("Nombre de voyages")
plt.xlabel("Catégorie")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Exemple de structure de DataFrame pour +=
data_plus_equal = {
    'Categorie': ['+=0', '+=Flex', '+=-'],
    '2023_Nombre de voyages': [1359050, 186405, 7039],
    '2024_Nombre de voyages': [2289521, 919552, 40948],
    '2023_Distance moyenne en km': [40, 45, 47],
    '2024_Distance moyenne en km': [40, 43, 46],
    '2023_Distance totale annuelle en km': [54362000, 8388225, 330833],
    '2024_Distance totale annuelle en km': [91580840, 39540736, 1883608],
}

df_plus_equal = pd.DataFrame(data_plus_equal)

# Tracer le nombre de voyages pour la catégorie += et ses sous-catégories
df_plus_equal.set_index('Categorie')[['2023_Nombre de voyages', '2024_Nombre de voyages']].plot(kind='bar', figsize=(10,6))
plt.title("Nombre de voyages pour += et ses sous-catégories (2023 et 2024)")
plt.ylabel("Nombre de voyages")
plt.xlabel("Catégorie")
plt.xticks(rotation=45)
plt.show()

# Tracer la distance moyenne en km pour la catégorie += et ses sous-catégories
df_plus_equal.set_index('Categorie')[['2023_Distance moyenne en km', '2024_Distance moyenne en km']].plot(kind='bar', figsize=(10,6))
plt.title("Distance moyenne en km pour += et ses sous-catégories (2023 et 2024)")
plt.ylabel("Distance moyenne (km)")
plt.xlabel("Catégorie")
plt.xticks(rotation=45)
plt.show()

# Tracer la distance totale annuelle en km pour la catégorie += et ses sous-catégories
df_plus_equal.set_index('Categorie')[['2023_Distance totale annuelle en km', '2024_Distance totale annuelle en km']].plot(kind='bar', figsize=(10,6))
plt.title("Distance totale annuelle en km pour les différentes tranches d'age (2023 et 2024)")
plt.ylabel("Distance totale annuelle (km)")
plt.xlabel("Catégorie")
plt.xticks(rotation=45)
plt.show()
```
