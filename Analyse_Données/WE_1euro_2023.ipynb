```python
import pandas as pd

# Chemin du fichier Excel
file_path = '/content/WE 1 2023.xlsx'

# Charger le fichier Excel
xls = pd.ExcelFile(file_path)

# Obtenir les noms des feuilles
sheet_names = xls.sheet_names

# Dictionnaire pour stocker les données des premières lignes de chaque feuille
first_fifteen_rows = {}

# Parcourir chaque feuille et afficher les 15 premières lignes
for sheet in sheet_names:
    df = pd.read_excel(xls, sheet_name=sheet)
    first_fifteen_rows[sheet] = df.head(15)  # Stocker les 15 premières lignes

    # Afficher les 15 premières lignes
    print(f"Premières 15 lignes pour {sheet}:")
    display(first_fifteen_rows[sheet])  # Utiliser 'display' pour un affichage plus propre dans des notebooks

```

```python
def extract_destination(parcours):
    if isinstance(parcours, str):  # Vérifie si la valeur est une chaîne de caractères
        return parcours.split(' - ')[-1].strip()
    else:
        return "Inconnu"  # Retourner une valeur par défaut ou NaN pour les non-chaînes
```

```python
for sheet in sheet_names:
    df = pd.read_excel(xls, sheet_name=sheet)
    df['Destination'] = df['Parcours'].apply(extract_destination)

    # Calculer la fréquence des destinations
    destination_counts = df['Destination'].value_counts()

    # Trouver la destination la plus fréquentée
    most_frequent_destination = destination_counts.idxmax()
    frequency = destination_counts.max()

    # Stocker dans le dictionnaire
    most_frequent_destinations[sheet] = (most_frequent_destination, frequency)

# Afficher les résultats
for month, (destination, count) in most_frequent_destinations.items():
    print(f"{month}: La destination la plus fréquentée est {destination} avec {count} voyages.")

```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Supposons que most_frequent_destinations soit déjà rempli avec les données nécessaires
months = list(most_frequent_destinations.keys())
destinations = [info[0] for info in most_frequent_destinations.values()]
counts = [info[1] for info in most_frequent_destinations.values()]

# Création d'un DataFrame pour faciliter la visualisation
visualization_df = pd.DataFrame({
    'Month': months,
    'Destination': destinations,
    'Count': counts
})
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Supposons que visualization_df soit déjà défini et contient les données nécessaires
plt.figure(figsize=(14, 10))
bars = plt.bar(visualization_df['Month'], visualization_df['Count'], color='skyblue')
plt.xlabel('Mois', fontsize=14)
plt.ylabel('Nombre de Voyages', fontsize=14)
plt.title('Destination la Plus Fréquentée par Mois', fontsize=16)
plt.xticks(rotation=45, fontsize=12)  # Incliner les noms des mois pour une meilleure visibilité

# Supposer que 'Toulouse Matabiau' est la destination fréquente pour tous les mois
destination_constante = 'Toulouse Matabiau'

# Ajouter une annotation ou une légende pour la destination
plt.figtext(0.5, 0.002, f'Destination la plus fréquentée pour tous les mois: {destination_constante}', ha='center', va='bottom', fontsize=12, color='darkred')

plt.tight_layout()
plt.show()


```

```python
import pandas as pd

# Chemin du fichier Excel
file_path = '/content/WE 1 2023.xlsx'

# Charger le fichier Excel
xls = pd.ExcelFile(file_path)

# Boucle pour traiter chaque feuille (mois)
most_traveled_routes = {}
for sheet in xls.sheet_names:
    df = pd.read_excel(xls, sheet_name=sheet)

    # Assurer que 'Nombre de voyages' est numérique
    df['Nombre de voyages'] = pd.to_numeric(df['Nombre de voyages'], errors='coerce')

    # Filtrer pour exclure les lignes où 'Parcours' est 'Total' ou toute autre entrée non désirée
    df = df[df['Parcours'].apply(lambda x: isinstance(x, str) and not x.lower().startswith('total'))]

    # Grouper par 'Parcours' et calculer le total des voyages
    total_travels = df.groupby('Parcours')['Nombre de voyages'].sum()

    # Trouver le parcours avec le total le plus élevé
    most_traveled_route = total_travels.idxmax()
    total_voyages = total_travels.max()

    # Stocker le parcours le plus fréquenté et le total des voyages dans le dictionnaire
    most_traveled_routes[sheet] = (most_traveled_route, total_voyages)

# Afficher les parcours les plus fréquentés pour chaque mois
for month, (route, voyages) in most_traveled_routes.items():
    print(f"{month}: Le parcours le plus fréquenté est {route} avec un total de {voyages} voyages.")
```

```python
# Initialiser la variable pour le total global
total_global = 0

# Dictionnaire pour stocker le total de chaque mois
monthly_totals = {}

# Boucle pour traiter chaque feuille (mois)
for sheet in xls.sheet_names:
    df = pd.read_excel(xls, sheet_name=sheet)

    # Identifier la ligne 'Total'
    total_row = df[df['Parcours'].str.lower() == 'total']['Nombre de voyages']

    # Vérifier si la ligne 'Total' existe et sommer
    if not total_row.empty:
        monthly_total = total_row.values[0]  # Assumer que 'Total' est un nombre unique dans la ligne
        monthly_totals[sheet] = monthly_total
        total_global += monthly_total

# Afficher le total global et les totaux mensuels
print(f"Total Global: {total_global}")
for month, total in monthly_totals.items():
    print(f"{month}: Total de voyages = {total}")
```

```python
# Calculer et afficher les pourcentages de chaque mois
for month, total in monthly_totals.items():
    percentage = (total / total_global) * 100  # Calcul du pourcentage
    print(f"{month}: {percentage:.2f}% du total global")
```

```python
import matplotlib.pyplot as plt

# Supposons que monthly_totals soit un dictionnaire contenant les totaux de chaque mois
months = list(monthly_totals.keys())
totals = list(monthly_totals.values())

plt.figure(figsize=(10, 5))
plt.plot(months, totals, marker='o')  # Ligne avec points pour chaque mois
plt.title('Tendance des Voyages par Mois')
plt.xlabel('Mois')
plt.ylabel('Nombre Total de Voyages')
plt.grid(True)
plt.xticks(rotation=45)  # Incliner les étiquettes des mois pour une meilleure visibilité
plt.tight_layout()
plt.show()
```
